Antes de comenzar, revisamos las variables categóricas para ver si todas son útiles, o podemos eliminar alguna. 
Después de nuestro análisis, decidimos que las variables categóricas como el nombre no son utiles ya que solo la ligan información de una persona con su nombre y no sirven mucho en el modelo

Primero observamos que hay pasajeros en el titanic que tienen un id con los primeros iguales, al juntarlos obsservamos que tienen caracetristicas parecidas, no en todos los casos, pero si en la mayoría, como la cabina, si estaban en sueño criogenico, o si eran VIP o no. Como no hay un patrón en el gasto de cada pasajero vamos a llenar los datos vacios de gastos directo a cero, ya que es la moda de cada columna numérica.

```python
#unicos_val = df2["PassengerId"].unique()
cat = ["RoomService","FoodCourt","Spa","VRDeck","ShoppingMall"]
df2[cat] = df2[cat].fillna(0.0)
```

En el caso de las variables categoricas que no son CryoSleep y Edad, se uso el identificador de cuatro números de las ID, para juntarlos en usuarios con le mismo identificador y llenar los vacios usando la moda de cada conjunto en cada columna.

```python
#Creamos la variable ids, la cual contiene los valores de 'PassengerId' de nuestro df2 
#Pero almacenamos estos valores como un arreglo de numpy
ids = df2['PassengerId'].to_numpy()

#Creamos la variable unicos_val, la cual contiene los valores únicos que hay en 'PassengerId' 
unicos_val = df2["PassengerId"].unique()

#Iteramos de cero a la longitud de la variable unicos_val
for j in range(len(unicos_val)):
    #Creamos una variable df_i, la cual contiene los valores únicos de la columna "PassengerId" 
    #Estos valores los busca de acuerdo al valor de la posición j de la lista unicos_val
    #Todos los valores de "PassengerId" que tienen el valor j
    df_i = df2[ df2["PassengerId"] == unicos_val[j] ]

    cat = ["HomePlanet","Cabin","Destination","VIP"]

    # df_i.shape[0] -dimensión del arreglo 
    rows = df_i.shape[0]
    
    #Hacemos una segunda iteración dentro de las variables de la lista cat
    for i in cat:
        #Creamos una variable que nos de la moda de la columna i en nuestro df_i
        moda = df_i[i].mode()
        try:
            #Le decimos que llene los valores vacios de la respectiva columna con la moda 
            df_i[i] = df_i[i].fillna(moda[0])
        except:
            #*++++++++************************++++++++************+++++++++++*************************+++++++++++++
            df_i[i] = df_i[i].fillna(df2[i].mode()[0])

    df2[df2["PassengerId"] == unicos_val[j]] = df_i
```

Luego con la información obtenida anteriormente, podemos llenar los vacios de Edad y CryoSleep de manera rápida, primero extraemos las filas donde no se gasto nada en ninguna de las variables numéricas. Al ya tenerlas, llenamos los vacios de CryoSleep con false, ya que gente que no gasto estaba toda dormida, en el caso de los vacios los llenamos con la moda de las edades de personas menores a 13 años.

```python
df_i = df2.loc[(df2["RoomService"] == 0.0) & (df2["FoodCourt"] == 0.0) & (df2["Spa"] == 0.0) & (df2["VRDeck"] == 0.0) & (df2["ShoppingMall"] == 0.0)]
print(df_i[df_i["Age"] > 12].shape)
print(df_i[df_i["Age"] <= 12].shape)
print(df_i.shape)
edad_med = df_i[df_i["Age"] < 13.0]
moda = int(edad_med["Age"].mode()[0])
df_i["CryoSleep"] = df_i["CryoSleep"].fillna(True)
df_i["Age"] = df_i["Age"].fillna(moda)
print(df_i.isna().sum())
df2.loc[(df2["RoomService"] == 0.0) & (df2["FoodCourt"] == 0.0) & (df2["Spa"] == 0.0) & (df2["VRDeck"] == 0.0) & (df2["ShoppingMall"] == 0.0)] = df_i
```

```python
#Confirmamos que ya no haya NANS en las variables categóricas
df2.isna().sum()
```

Luego lo que hacemos es obtener las filas donde si se gasto algo, luego sabiendo esto llenamos los vacios de CryoSleep con False, y los vacios de edad con la moda de edad de gente mayor o igual a 13 años.

```python
df_i = df2.loc[(df2["RoomService"] != 0.0) | (df2["FoodCourt"] != 0.0) | (df2["Spa"] != 0.0) | (df2["VRDeck"] != 0.0) | (df2["ShoppingMall"] != 0.0)]
print(df_i[df_i["Age"] > 12].shape)
print(df_i[df_i["Age"] <= 12].shape)
print(df_i.shape)
edad_med = df_i[df_i["Age"] >= 13.0]
moda = int(edad_med["Age"].mode()[0])
df_i["CryoSleep"] = df_i["CryoSleep"].fillna(False)
df_i["Age"] = df_i["Age"].fillna(moda)
print(df_i.isna().sum())
df2.loc[(df2["RoomService"] != 0.0) | (df2["FoodCourt"] != 0.0) | (df2["Spa"] != 0.0) | (df2["VRDeck"] != 0.0) | (df2["ShoppingMall"] != 0.0)] = df_i
```

Al ya tener todo lleno podemos analizar la columna de cabina, la cual tiene una letra al inicio y otra al final, lo que hacemos es extraer estas dos letras y eliminar el número de en medio.

### Transformacion de variable categóricas a numéricas

Como vimos anteriormente hay dos variables con solo dos categorías, que son CryoSleep y VIP. Usaremos label encoder en esas dos variables para que sean 0 o 1. En las demás variables hay muchas categorías por lo que usaremos dummy variables.

```python
df2.dtypes
```

```python
for i in df2:
    if df2[i].dtype == 'object':
        print(df2[i].value_counts())
```

Usamos el laber encoder para poder cambiar las variables con solo dos categorías

```python
label_encoder = LabelEncoder()
le_cat = ["CryoSleep","VIP","Transported","tipo_cabin"]
for i in le_cat:
    df2[i]=label_encoder.fit_transform(df2[i]) 
```

```python
df2.head()
```

Para las variables con más de dos categorías usamos dummy variables para poder agregar una columna por cada una de las categorías.

```python
dummy_cat = ["HomePlanet","Destination","letra_cabin"]
for i in dummy_cat:
    dummies = pd.get_dummies(df2[i])
    df2 = pd.concat([df2.drop(columns=[i]), dummies], axis=1)
```

Al final obtenemos el dataset final y está listo para poder entrenar el modelo

### Escalamiento

```python
#Para escalar los valores trabajamos solo con valores numéricos 
#Escalamos las variables con la escala de min-max
#Los valores estandarizados con la escala de min-max caen en el mismo intervalo [0,1], pero no tienen la misma 
#media ni desviación estandar 
cat = ["RoomService","FoodCourt","Spa","VRDeck","ShoppingMall","Age"]
for i in cat:
    df2[i] = (df2[i]-df2[i].min())/ (df2[i].max()-df2[i].min())
```
